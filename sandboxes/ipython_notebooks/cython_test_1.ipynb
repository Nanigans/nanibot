```python
%%cython --annotate
import time

def f(double x):
    return x**2-x

def integrate_f(double a, double b, int N):
    cdef int i
    cdef double s, dx
    s = 0
    dx = (b-a)/N
    for i in range(N):
        s += f(a+i*dx)
    return s * dx

start = time.time()
result = integrate_f(0, 10, 10000)
stop = time.time()

print result
print 'duration: {}'.format(stop-start)
```
